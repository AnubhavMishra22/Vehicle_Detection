```python
import cv2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
car_cascade = cv2.CascadeClassifier('cars.xml')
```

```python
frame = cv2.imread('./images/image (1).jpg')
```

```python
# Image Input
plt.imshow(frame)
plt.show()
```

```python
gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
```

```python
cars = car_cascade.detectMultiScale(gray, 1.1, 1)
```

```python
for (x,y,w,h) in cars:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
```

```python
plt.imshow(frame)
plt.show()
```

```python
# Number of Vehicles(cars) detected
print(cars.shape[0])
```
